SMT (Statistical Machine Translation) — статистическая машинная переводческая система, использующая параллельные корпуса текстов для обучения языковым моделям. Основная цель SMT — предсказать наиболее вероятный перевод фразы на основе вероятностных моделей.

Компоненты SMT:
Параллельные корпуса- двухъязычные тексты, где исходные предложения соответствуют переведенным предложениям.

Выравнивание - процесс сопоставления слов или фраз в исходном тексте и их переводе.
Виды выравнивания: слово-слово, фразовое.
Языковая модель

Вероятностная модель, оценивающая, насколько текст грамматически правильный и естественный в языке перевода.
Чаще всего основана на n-граммах (последовательностях из n слов)

Модель перевода - вероятностная модель, описывающая, как слова и фразы из исходного языка преобразуются в слова и фразы языка перевода.

Декодер - алгоритм, который выбирает наиболее вероятный перевод на основе моделей перевода и языковой модели.

Tiny SMT
Основные этапы работы:
Препроцессинг:токенизация, очистка данных, нормализация текста.
Формирование параллельных корпусов.
Выравнивание: используются модели, такие как IBM Model 1, Model 2, или алгоритм GIZA++.
Обучение языковой модели: построение n-граммной языковой модели с использованием библиотек например, NLTK
Обучение модели перевода- вычисление вероятностей перевода фраз из одного языка в другой.
Декодирование- генерация перевода для новых текстов на основе обученных моделей.


NLTK: библиотека для обработки текста и построения языковых моделей.
SpaCy: библиотека для токенизации, нормализации и других NLP-задач.

Scikit-learn: используется для построения вероятностных моделей.
