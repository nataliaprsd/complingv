1. **Выбранная пара языков**: русский - испанский

2. **Пре-процессинг:**
   - **Токенизация**: библиотека NLTK для русского языка, библиотека spaCy или Stanza для испанского языка (разделение текста на слова и знаки препинания);
   - **Нормализация**: NLTK для русского языка, spaCy для испанского (удаление лишних пробелов, преобразование текста в нижний регистр);
   - **Лемматизация**: pymorphy2 для русского языка, spaCy для испанского (приведение слов к их начальной форме).

3. **Примеры правил для языковой пары русский - испанский:**
   - **Правило №1:** Прямой перевод существительного. Для русского языка учитываются род, число и падеж, для испанского языка — род и число. Например, 'кот' (рус.) переводится как 'el gato' (исп.), где 'el' — артикль для мужского рода.
   - **Правило №2:** Прямой перевод прилагательного. Для русского языка учитываются род, число и падеж, для испанского языка — род и число. Например, 'красивый дом' (рус.) переводится как 'la casa bonita' (исп.), с учетом порядка слов.
   - **Правило №3:** Перевод возвратных глаголов. В испанском языке возвратные глаголы требуют использования местоимения 'se', тогда как в русском языке используется возвратная частица '-ся'. Например, 'мыться' (рус.) переводится как 'lavarse' (исп.).
